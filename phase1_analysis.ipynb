```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.tseries.offsets import *
%matplotlib inline
```

```python
train_data2 = pd.read_csv("./dataSet_phase2/dataSet_phase2/volume(table 6)_training2-1.csv")
train_data2.head()
```

```python
train_data2.info()
```

```python
volume_data = train_data2[["date_time", "tollgate", "direction"]].copy()
volume_data = volume_data.sort_values(by=["date_time"])
volume_data["date_time"] = volume_data["date_time"].apply(pd.Timestamp)
tollgate_direction = [[1, 0], [1, 1], [2, 0], [3, 0], [3, 1]]
# tollgate_direction = [[1, 0]]
result = pd.DataFrame()

def generate_feature(volume, tollgate, direction):
    temp = volume.copy()
    temp["tollgate_id"] = tollgate
    temp["direction"] = direction
    temp["time_window"] = temp["time"].apply(lambda time_basic: "[" + str(time_basic) + "," + \
                                                  str(time_basic + DateOffset(minutes=20)) + ")")
    temp["time"]
    return temp
    
for tollgate, direction in tollgate_direction:
    temp = volume_data[(volume_data["tollgate"] == tollgate) & (volume_data["direction"] == direction)][["date_time"]]
    temp["volume"] = 1
    temp.index = temp["date_time"]
    del temp["date_time"]
#     print(temp)
    temp = temp.resample("20T").sum()
    
    # 取8点到10点，17点到19点的数据
    temp["time"] = temp.index
    temp["hour"] = temp["time"].apply(lambda x: x.hour)
    temp1 = temp[(temp["hour"] >= 8) & (temp["hour"] < 10)][["volume", "time"]]
    temp1 = generate_feature(temp1, tollgate, direction)
    temp2 = temp[(temp["hour"] >= 17) & (temp["hour"] < 19)][["volume", "time"]]
    temp2 = generate_feature(temp2, tollgate, direction)
    result = result.append(temp1, ignore_index=True)
    result = result.append(temp2, ignore_index=True)

result
```

```python
predict_result = pd.read_csv("./training/volume_predict_multi_sample15_filter.csv")
predict_result.head()
```

```python
result_compare = pd.merge(predict_result, result, on=["tollgate_id", "time_window", "direction"])
result_compare
```
